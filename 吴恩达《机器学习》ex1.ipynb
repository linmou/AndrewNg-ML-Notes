```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path =  'H:/吴恩达《机器学习》/andrew_ml_ex14179/ex1data1.txt'
data = pd.read_csv(path, header=None, names=['Population', 'Profit'])
data.head()
```

```python
data.plot(kind='scatter', x='Population', y='Profit', figsize=(12,8))
plt.show()
```

```python
data.insert(0,'ONEs',1)
X=data.iloc[:,0:-1]
y=data.iloc[:,-1]
theta=np.zeros(X.shape[-1])
```

```python
def computeCost(X,y,theta):
    cost=np.power((np.mat(X)*np.mat(theta).T-np.mat(y).T),2)
    return np.sum(cost)/(2*len(X))
```

```python
def GradientReg(X,y,theta,lr,iters):
#     x_para=X.shape[1]
    cost=np.zeros(iters)
    for i in range(iters):
        times=np.mat(X)*np.mat(theta).T-np.mat(y).T
        gradiant=times.T*np.mat(X)/len(X)
        theta =theta - lr*gradiant
        cost[i] = computeCost(X,y,theta)
    return theta ,cost
```
