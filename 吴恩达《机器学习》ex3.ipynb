```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from scipy.io import loadmat
from sklearn.metrics import classification_report

data = loadmat(r'H:\吴恩达《机器学习》\andrew_ml_ex33507\ex3data1.mat')
data

X=data['X']
X=np.insert(X,0,np.ones(X.shape[0]),axis=1)
y=data['y']
theta=np.zeros((X.shape[1],10))
learning_rate=1
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def  cost(theta,X,y):
    y=np.mat(y).reshape(-1,1)
    theta=np.mat(theta).reshape(X.shape[1],-1)
    h=sigmoid(X@theta)
    return (-y.T@np.log(h)-(1-y).T@(np.log(1-h)))/len(X)

def gradient(theta,X,y):
    y=np.mat(y).reshape(-1,1)
    theta=np.mat(theta).reshape(X.shape[1],-1)
    return (sigmoid(X@theta)-y).T@X/len(X)

def costReg(theta, X, y, learningRate):
    X = np.matrix(X)
    theta=np.mat(theta).reshape(X.shape[1],-1)
    y = np.matrix(y)
    theta1=theta[1:,:]
    reg = (learningRate / (2 * len(X))) * (theta1.T@theta1)
    return cost(theta, X, y)+ reg

def gradientReg(theta, X, y, learningRate):
    X = np.matrix(X)
    theta=np.mat(theta).reshape(X.shape[1],-1)
    y = np.matrix(y)
    grad= np.zeros(theta.shape[1])
    grad=gradient(theta, X, y)+learningRate*(theta.T)/(len(X))
    grad[0,0]=np.sum(np.multiply(sigmoid(X * theta) - y, X[:,0])) / len(X)
    return grad.ravel()

def prediction(theta,X):
    X = np.matrix(X)
    theta=np.mat(theta).reshape(X.shape[1],-1)
    probability = sigmoid(X * theta)
    return [1 if x >= 0.5 else 0 for x in probability]
```

```python
def predict_all(X, theta):
    # convert to matrices
    X = np.matrix(X)
    theta = np.matrix(theta).reshape(X.shape[1],-1)
    
    # compute the class probability for each class on each training instance
    h = sigmoid(X *theta)
    
    # create array of the index with the maximum probability
    h_argmax = np.argmax(h, axis=1)
    
    # because our array was zero-indexed we need to add one for the true label prediction
    h_argmax = h_argmax + 1
    
    return h_argmax
```

```python
y_pred = predict_all(X, theta)
print(classification_report(y, y_pred))
```
